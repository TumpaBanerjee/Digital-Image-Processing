# Calculate frequency of intensities

```python
p={}
for i in range(256):
    p[i]=0
print(p)
```

```python
height,width=img1.shape
for row in range(width):
    for column in range(height):
        intensity=img1[column][row]
        p[intensity]=p[intensity]+1
print(p)
```

# probability density function for intensities

```python
pdf=copy.deepcopy(p)
for i in pdf:
    pdf[i]=pdf[i]/(height*width)
print(pdf)
```

# cumulative density function 

```python
cdf=copy.deepcopy(pdf)
cum_int=0
for i in cdf:
    cum_int+=cdf[i]
    cdf[i]=cum_int
print(cdf)
```

# Discretize the intensitiy values

```python
sk=copy.deepcopy(cdf)
for i in sk:
    sk[i]=sk[i]*255
print(sk)
```

```python
new_value=copy.deepcopy(sk)
for i in new_value:
    new_value[i]=round(new_value[i])
print(new_value)
```

# Create image with new intensity values

```python
new_image=img1.copy()
height,width=new_image.shape
for i in range(height):
    for j in range(width):
        ints=new_image[i][j]
        #print(ints)
        #print(new_value[ints])
        new_image[i][j]=new_value[ints]
```
