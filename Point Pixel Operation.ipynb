```python
neg_image=np.array((img1))
print(neg_image.shape[0])
```

```python
def negative_image(img):
    neg_img=np.array(img)
    row=neg_image.shape[0]
    col=neg_image.shape[1]
    for i in range(row):
        for j in range(col):
            neg_img[i][j]=255-img[i][j]
    return neg_img
```

```python
neg_img=negative_image(img1)
```

```python
neg_img_rgb=cv2.cvtColor(neg_img,cv2.COLOR_BGR2RGB)
plt.imshow(neg_img_rgb)
```

```python
import math
def log_transformation(img2):
    print(img2.shape)
    log_img=np.array(img2)
    row1,col1=img2.shape
    #col=neg_image.shape[1]
    #print(row1,col1)
    #print(img2)
    for i in range(row1):
        for j in range(col1):
            log_img[i][j]=(math.log(1+img2[i][j]/255.0))*255.0
    log_img.astype('uint8')
    return log_img
```

```python
print(img2.shape)
log_img=log_transformation(img2)
```

```python
log_img
```

```python
log_img_rgb=cv2.cvtColor(log_img,cv2.COLOR_BGR2RGB)
plt.imshow(log_img_rgb)
```

```python

def gamma_transformation(img,gamma=0.6):
    print(img.shape)
    gamma_img=np.array(img)
    row,col=img.shape
    #gamma=0.6
    #col=neg_image.shape[1]
    print(row,col)
    #print(img2)
    for i in range(row):
        for j in range(col):
            gamma_img[i,j]=((img[i,j]/255.0)**gamma)*255.0
    gamma_img.astype('uint8')
    return gamma_img
```

```python
gamma_img=gamma_transformation(img3,gamma=0.6)
```

```python
gamma_img_rgb=cv2.cvtColor(gamma_img,cv2.COLOR_BGR2RGB)
plt.imshow(gamma_img_rgb)
```

```python
gamma_img=gamma_transformation(img3,gamma=0.4)
gamma_img_rgb=cv2.cvtColor(gamma_img,cv2.COLOR_BGR2RGB)
plt.imshow(gamma_img_rgb)
```

```python
gamma_img=gamma_transformation(img3,gamma=0.3)
gamma_img_rgb=cv2.cvtColor(gamma_img,cv2.COLOR_BGR2RGB)
plt.imshow(gamma_img_rgb)
```
