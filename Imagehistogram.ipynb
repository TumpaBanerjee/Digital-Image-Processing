```python
def imhist1(img):
    hist=np.zeros(256)
    for i in range(256):
        hist[i]=count=np.count_nonzero(img == i)
    return hist
```

```python
def imhist(img):
    hist=np.zeros((256))
    for i in range(256):
        hist[i]=sum(sum(img==i))
    return hist
```
