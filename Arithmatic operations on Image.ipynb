Arithmetic operations between two images f (x, y) and g(x, y) are denoted as
s(x,y)=f(x,y)+g(x,y)
d(x,y)=f(x,y)-g(x,y)

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# Read two images

```python
img1=cv2.imread('C:\MY DRIVE\Material\Digital Image Processing\Image database\Arithmatic Operations\image1.jpg',0)
img2=cv2.imread('C:\MY DRIVE\Material\Digital Image Processing\Image database\Arithmatic Operations\image2.jpg',0)
```

# Plot the images

```python
ax1 = plt.subplot(1,2,1)
plt.imshow(img1,cmap='gray')
ax1 = plt.subplot(1,2,2)
plt.imshow(img2,cmap='gray')
```

# Use cv2 library for image addition

```python
img_add=cv2.add(img1,img2)
plt.imshow(img_add,cmap='gray')
```

# Perform simple arithmatic addition

```python
img3=img1+img2
plt.imshow(img3,cmap='gray')
```
