```python
def contrast_stretching(img,r1=0,s1=0,r2=255,s2=255):
    imgT=np.array(img)
    row,col=img.shape
    s1=(r1-0)/(s1-0)
    s2=(r2-r1)/(s2-s1)
    s3=(255-r2)/(255-s2)
    print(row)
    print(col)
    for r in range(row):
        for c in range(col):
            if (img[r,c]>=0 and img[r,c]<=r1):
                imgT[r,c]=((img[r,c]/255.0)*s1)*255.0
            if (img[r,c]>=r1 and img[r,c]<=r2):
                imgT[r,c]=((img[r,c]/255.0)*s2)*255.0
            else:
                imgT[r,c]=((img[r,c]/255.0)*s3)*255.0
    imgT.astype('uint8')
    return imgT
```

```python
imgT1=contrast_stretching(img1,r1=50,s1=30,r2=110,s2=215)
imgT1_rgb=cv2.cvtColor(imgT1,cv2.COLOR_BGR2RGB)
plt.imshow(imgT1_rgb)
```

```python
imgT1=contrast_stretching(img1,r1=90,s1=20,r2=110,s2=230)
imgT1_rgb=cv2.cvtColor(imgT1,cv2.COLOR_BGR2RGB)
plt.imshow(imgT1_rgb)
```

```python
def intensity_level_slicing(img):
    imgT=np.array(img)
    row,col=img.shape
    print(row)
    print(col)
    for r in range(row):
        for c in range(col):
            if (img[r,c] >=142 and img[r,c] <=255):
                imgT[r,c]=255
            else:
                imgT[r,c]=30
    return imgT
```

```python
imgT2=intensity_level_slicing(img2)
```

```python
imgT2_rgb=cv2.cvtColor(imgT2,cv2.COLOR_BGR2RGB)
plt.imshow(imgT2_rgb)
```

```python
def intensity_level_slicing_2(img):
    imgT=np.array(img)
    row,col=img.shape
    print(row)
    print(col)
    for r in range(row):
        for c in range(col):
            if (img[r,c] >=142 and img[r,c] <=255):
                imgT[r,c]=200
            else:
                imgT[r,c]=img[r,c]
    return imgT
```

```python
imgT3_rgb=cv2.cvtColor(imgT3,cv2.COLOR_BGR2RGB)
plt.imshow(imgT3_rgb)
```

```python
def bit_slicing(img,ind):
    s=[]
    row,col=img.shape
    for r in range(row):
        for c in range(col):
            s.append(np.binary_repr(img[r,c],width=8))
    bit_img=(np.array([int(i[ind]) for i in s],dtype=np.uint)*1).reshape(row,col)
    bit_img.astype('uint8')
    return bit_img
    
```
